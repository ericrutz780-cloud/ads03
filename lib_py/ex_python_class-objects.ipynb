# Objekten und Klassen

```python
import inspect
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as plt
```

## Datensatz laden

```python
# Aus dem Titanic-Beispiel
titanic = sns.load_dataset("titanic")
titanic.head()
```

## Funktionen

```python
def age_range(a):
    if a["age"] <= 20:
        return "young"
    elif a["age"] >= 20:
        return "older"
    elif np.isnan(a["age"]):
        return np.NaN
    else:
        return "age unknown"
```

```python
age_range(35)
```

```python
titanic.apply(age_range, axis=1)
```

```python
titanic["age_ranges"] = titanic.apply(age_range, axis=1)
```

```python
titanic.head(5)
```

```python
titanic["age_ranges"].value_counts().plot.bar()
```
