### Data Extraction

#### Scraper Funktion

```python
def scrape_website(name, url):
    # (1) Run request
    response = requests.get(url, allow_redirects=True)
    content = response.content
    text = response.text

    # (2) File name to store the raw HTML
    file_name = os.path.join(
        STORAGE_DIR,
        f"{now_str}-{name}.html",
    )

    # (3) Write raw HTML
    with open(file_name, "wb") as f:
        f.write(response.content)

    # (4) Fill content_dict and text_dict
    content_dict[name] = response.content
    text_dict[name] = response.text

    # (5) Fill log_list
    log_info = dict(
        name=name,
        date=now_str,
        file_name=file_name,
        status=response.status_code,
        url=response.url,
        encoding=response.encoding,
    )
    log_list.append(log_info)
```

**AusfÃ¼hren**

```python
for name, url in newspaper_urls.items():
    try:
        scrape_website(name, url)
    except:
        failing_list.append((name, url))
```

### Liste sÃ¤ubern

```python
def list_cleaner(liste):

    list_cleaning = ['{','"', '}']

    json_data = []

    for element in liste:
        temp_element = element
    
        for delete in list_cleaning:
            temp_element = temp_element.replace(delete, "")
        
        json_data.append(temp_element)

    return json_data
```
